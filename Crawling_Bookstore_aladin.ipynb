```python
# ライブラリのImport
from selenium import webdriver
from bs4 import BeautifulSoup
import time
import codecs
```

```python
# Chromedriverのセッティング
driver = webdriver.Chrome("C:/Users/dltjs/Data20190921/day3/chromedriver.exe")

# 書店のベストセラーページに移動
driver.get("https://www.aladin.co.kr/shop/common/wbest.aspx?BranchType=1&start=we")

# ソースコードを持ってくる.
source = driver.page_source
```

```python
# parsing
s1 = BeautifulSoup(source, "html.parser")

# div tag ながら class が ss_book_box である要素の数を確認する.
s_book = s1.find_all("div", class_="ss_book_box")
len(s_book)
```

```python
# s_book構造を把握する。
for li in s_book[0].find_all("li"):
    print(li.text)
    print("-"*70)

# 1、2、3番目の要素だけが出力してみる。
print(s_book[0].find_all("li")[1].text)
print(s_book[0].find_all("li")[2].text)
print(s_book[0].find_all("li")[3].text)
```

```python
# s_book 繰り返し処理すること.
for item in s_book: # 順位別に持ってくること
    for li in item.find_all("li"): # 順位でliだけを持ってくる.
        print(li.text)
    print("="*70)
```

```python
# s_book繰り返し処理すること->出力をConsoleではなくresult2.txtとする.
for item in s_book:
    li = item.find_all("li")
    if str(li[0].text)[0] != '[': # プレゼントがない場合
        print(li[0].text)
        print(li[1].text)
        print(li[2].text)
    else: #プレゼントがある場合
        print(li[1].text)
        print(li[2].text)
        print(li[3].text)
    print("="*70)
```

```python
# s_book "li" 要素長を順位別に出力してくれる.
rank = 0
for item in s_book:
    print(str(rank+1)+"위 책 li 길이")
    print(len(item.find_all("li")))
    rank += 1
```

```python
f = codecs.open("C:/Users/dltjs/Data20190921/day4/result2.txt", encoding="utf-8", mode='w')
# s_book繰り返し処理すること->出力をConsoleではなくresult2.txtとする.
rank = 0
for item in s_book:
    li = item.find_all("li")
    rank += 1
    f.write("# "+str(rank)+"위 도서 \r\n")
    if str(li[0].text)[0] != '[':
        f.write(li[0].text+"\r\n")
        f.write(li[1].text+"\r\n")
        f.write(li[2].text+"\r\n")
    else:
        f.write(li[1].text+"\r\n")
        f.write(li[2].text+"\r\n")
        f.write(li[3].text+"\r\n")
    f.write("="*70 + "\r\n")
f.close()
```
