```python
# ライブラリインポート
from selenium import webdriver
from bs4 import BeautifulSoup
import time
import codecs
```

```python
# Chromedriverのセッティング
driver = webdriver.Chrome("C:/Users/dltjs/Data20190921/day3/chromedriver.exe")

# 書店のベストセラーページに移動
driver.get("http://www.yes24.com/24/Category/BestSeller")

# ソースコードを持ってくる
source = driver.page_source
```

```python
# parsing
s1 = BeautifulSoup(source, "html.parser")

#"ol"の要素の数を確認する.
s_book = s1.find_all("ol")
print(len(s_book))
```

```python
len(s_book[0].find_all("li", class_=["num1", "num2", "num3"]))
```

```python
# 下のリストに .append()を利用して
# forを40回繰り返し、最初の刺身には"num"+str(1)のように"num40"までappendさせする、反復を作成。
class_list = []

for i in range(1, 41):
    if i == 19 or i == 20:
        class_list.append("num"+str(i)+"_line")
    else:
        class_list.append("num"+str(i))
print(class_list)
```

```python
len(s_book[0].find_all("li", class_=class_list))
```

```python
#上のコードを応用して順位別に本のタイトルー>作家ー>価格順に並んでください.
rank = 0
for item in s_book[0].find_all("li", class_=class_list):
    p = item.find_all("p")
    rank += 1
    print("# "+str(rank)+"위 도서")
    print("# 제목 : "+p[2].text)
    print("# 저자 : "+p[3].text)
    print("# 가격 : "+p[4].text)
    print("="*70)
```

```python
#day4フォルダの下にresult3.txtというファイルを作ってください.
f = codecs.open("C:/Users/dltjs/Data20190921/day4/result3.txt", encoding="utf-8", mode='w')

# s_book繰り返し処理すること->出力をConsoleではなくresult3.txtとする.
rank = 0
for item in s_book[0].find_all("li", class_=class_list):
    p = item.find_all("p")
    rank += 1
    f.write("# "+str(rank)+"위 도서 \r\n")
    f.write("# 제목 : "+p[2].text+"\r\n")
    f.write("# 저자 : "+p[3].text+"\r\n")
    f.write("# 가격 : "+p[4].text+"\r\n")
    f.write("="*70+"\r\n")
f.close()
```
