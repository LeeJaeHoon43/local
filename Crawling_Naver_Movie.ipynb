```python
# ライブラリインポート
from selenium import webdriver
from bs4 import BeautifulSoup
import time
import codecs
```

```python
# Chromedriverのセッティング.
driver = webdriver.Chrome("C:/Users/dltjs/Data20190921/day3/chromedriver.exe")

# Naverの映画サイトに移動.
driver.get("https://movie.naver.com/movie/running/premovie.nhn#")

# ソースコードを持ってくる
source = driver.page_source
```

```python
# parsing
m1 = BeautifulSoup(source,"html.parser")

#ソースコードm1でTagがdtで、classがtitであることを出力してみて、要素の数を確認する.
s_movie = m1.find_all("dt",class_="tit")
print(s_movie)
len(s_movie)
```

```python
# s_movieの構造を確認する。
for name in s_movie[0].find_all("a"):
    print(name.text.strip())
    print("="*70)
```

```python
# s_movie繰り返し処理する.
list_name = []
for item in s_movie:
    for li in item.find_all("a"):
        print(li.text.strip().replace("\t",""))
        list_name.append(li.text.strip().replace("\t",""))
    print("="*70)
print(list_name)
```

```python
#ソースコードm1でTagがdlであり、classがinfo_txt1であることを出力してみて、要素の数を確認する.
s_info = m1.find_all("dl",class_="info_txt1")
print(s_info)
len(s_info)
```

```python
# s_infoの構造を確認する。
for name in s_info:
    print(name.text.strip().replace("\t","").replace("\n","").replace("개요","#장르 : ").replace("감독","\n#감독 : ").replace("출연","\n#출연 : "))
    print("="*70)
```

```python
# s_info繰り返し処理する。
list_info = []
for name in s_info:
    list_info.append(name.text.strip().replace("\t","").replace("\n","").replace("개요","#장르 : ").replace("감독","\n#감독 : ").replace("출연","\n#출연 : "))
print(list_info)
```

```python
print("***개봉 예정 영화***")
print("="*70)
for idx in range(len(list_name)):
    print("#제목 :",list_name[idx])
    print(list_info[idx])
    print("="*70)
```
