## 1) ギョルチュクチがあるレコードを削除

- 欠測値を確認
- 一つでもあるデータを削除
- すべての値が欠測値のデータを削除

```python
# Columnの欠測値の数 : .isnull.sum()
data.isnull().sum()
```

```python
# 欠測値のあるRowだけを残して照会する: .isnull()。any(axis=1)
data[data.isnull().any(axis=1)]
```

```python
# 欠測値のあるRowのインデックスの番号だけを抽出する。
nan_idx = data[data.isnull().any(axis=1)].index
nan_idx
```

```python
#一つでも欠測値があるレコードを削除 : .dropna()
# data : 24 x 7
# 21 x 7
data.dropna()
```

```python
# 削除確認:.dropna().isnull().sum()
data.dropna().isnull().sum()
```

```python
# すべての値が欠測値のレコードを削除: .dropna(how='all')
# data(全体):7つのColumnが全部ギョルチュクチのユヒョンマン削除する.
data.dropna(how='all')
```

```python
# 二つのColumn('、一般細菌'、'窒酸性窒素')が欠測値のRowを削除する。 二重のリストでまとめて[], dropna(how='all')
# 全体DataFrameにすぐ適用できない
data[['일반세균', '질산성질소']].dropna(how='all')
```

```python
# '一般細菌', '硝酸性酸素' Columnをコピー.
tmp = data[['일반세균', '질산성질소']].copy()
# 一つでも欠測のあるレコードを持ってくる:条件索引の使用
tmp[tmp.isnull().any(axis=1)]
```

```python
# すべての値が欠測のあるレコードを持ってくる
tmp[tmp.isnull().all(axis=1)]
```

```python
# drop(idx/col、axis=0(row削除)、1(columns削除))
data.drop(16, axis=0)
```

```python
#欠測値(NaN値)を0に代替する。
data2 = data.fillna(0)
```

## 3) データの平均値を使用して欠測値代替

```python
# 欠測値(NULL値)をデータの平均値に置き換える。
# 全データの属性のタイプを確認し、数字データに対する平均値を使用する。
# 一般細菌、硝酸性...
# df.select_dtypes(include=特定タイプ) # int/float/object
# df._get_numeric_data()
data._get_numeric_data()
```

```python
avg1 = tmp.fillna(0).values.mean()
```

```python
avg1
```

```python
# 平均値でdata変数の欠測値を満たした後、欠測値があったRowだけを照会。
data.fillna(avg1).iloc[nan_idx]
```

## 4) 欠測値が属したClassの属性の平均値を使用して欠測値を代替する。

```python
data3 = data.copy()
```

```python
#一般細菌がnullのRowだけを選んで出力。
data['일반세균'][data['일반세균'].isnull()]
```

```python
# 一般細菌Columnの平均.
avg1 = data3['일반세균'].mean()
avg1
```

```python
# 一般細菌のColumnの欠測値をavg1で埋めてdata3に永久的に保存。
data3['일반세균'].fillna(avg1, inplace=True)
data3
```

```python
data3['일반세균'].iloc[[16, 19]]
```

```python
# data3のDataFrameの16、19のColumnだけを選んで出力。
data3['질산성질소'][data['질산성질소'].isnull()]
```

```python
avg2 = data3['질산성질소'].mean()
avg2
```

```python
data3['질산성질소'].fillna(avg2, inplace=True)
```

```python
data3['질산성질소'].iloc[[3, 16]]
```

```python
#適合というColumnの値が不適合という資料だけを出力する.
data[data['적합'] == '부적합']
```

```python
# 적합 컬럼이 적합인 자료만 모아보기
data[data['적합'] == '적합']
```

```python
# 適合の有無別に属性星の平均値を求め、欠測値を代替
# groupby() : 範疇型データについて集計基準を設定
data4 = data.copy()
```

```python
# 適合のColumnを基準に集計した「一般細菌」dの平均
avg3 = data4.groupby('적합')['일반세균'].mean()
avg3
```

```python
# pivot_table 使用
data4.pivot_table(index='적합', values='일반세균', aggfunc='mean')
```

```python
# avg3の0番目は不適合の平均、1番目は適合の平均
print(avg3[0], avg3[1])
```

```python
# 適合には適合平均、不適合には不適合平均大学入試
data4['일반세균'][data4['적합']=='적합'] = data4['일반세균'][data4['적합']=='적합'].fillna(avg3[1])
```

```python
data4['일반세균'][data4['적합']=='부적합'] = data4['일반세균'][data4['적합']=='부적합'].fillna(avg3[0])
```

```python
data4
```

```python
avg4 = data4.groupby('적합')['질산성질소'].mean()
avg4
```

```python
# avg4の0番目は不適合の平均、1番目は適合の平均
print(avg4[0], avg4[1])
```

```python
# 適合には適合平均を代入する
data4['질산성질소'][data4['적합']=='적합'] = data4['질산성질소'][data4['적합']=='적합'].fillna(avg4[1])
```

```python
#不適合には不適合平均を代入する
data4['질산성질소'][data4['적합']=='부적합'] = data4['질산성질소'][data4['적합']=='부적합'].fillna(avg4[0])
```
