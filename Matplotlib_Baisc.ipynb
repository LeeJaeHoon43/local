### スタイル指定

- plt.plot([x], y, [fmt], **kwargs)
    - fmt: '[color][marker][line]'
        - カラー、マーカー、線などに対するフォーマットオプションを設定

```python
# スタイル文字の追加 : [fmt] => color + maker + lineの順番に文字を作成
# 省略された要素はデフォルト値として適用する。
# o:ライン無しで大きな点(円形)でデータを表示
plt.plot([1, 2, 3, 4], [2, 6, 10, 7], 'o')
plt.show()
```

```python
# - : マーカーは丸、ラインは実線
# fmt [カラー][マーカー][線の種類]
plt.plot([1, 2, 3, 4], [2, 6, 10, 7], 'o-')
plt.show()
```

```python
# o-- : 마커는 동그라미, 선은 점선
plt.plot([1, 2, 3, 4], [2, 6, 10, 7], 'o--')
plt.show()
```

```python
# ro::r(赤色),o(マーカー),:(丸い点線)
plt.plot([1, 2, 3, 4], [2, 6, 10, 7], 'ro:')
plt.show()
```

```python
# その他のスタイルパラメータの種類
# color (線の色) : 色 (magenta)
# linestyle(線の形) : -(-.)
# marker(マーカーの形) : ^(三角形)
# makeredgecolor(マーカーの縁の色) : r(赤)
# markerfacecolor(マーカーの色) : orange
# markersize(マーカーの大きさ):20
plt.plot([1, 2, 3, 4], [2, 6, 10, 7], color='magenta', linestyle='-.',
        marker='^', markeredgecolor='r', markerfacecolor='orange', markersize='20')
```

### matplotlibpyplot.scatter

- 代表値を中心にデータが散らばっている程度を視覚化
- データの分散度合いとデータ間の相関が把握できる。
- plt.scatter(x, y, s=None, c=None, data=None)
    - s : マーカーサイズ
    - c : マーカー色
    - data: dictionary 資料型を使用してデータを伝達すれば keyword string を通じてデータの使用が可能

```python
# 正規分布確率密度で100の標本を抽出
# random.normal()
np.random.seed(123)
x = np.random.normal(0, 1, 100)
y = np.random.normal(0, 1, 100)
plt.scatter(x, y)
plt.show()
```

```python
# Dictionaryタイプでデータを配信
# サンプルの意味:データごとに色とマーカーの大きさが違うように表現されるために生成する。
# xデータ(a):0以上50未満の範囲で1ずつ増加する整数
# マーカーの色データ(b):0以上50未満の整数の中で50個の乱数を抽出
# マーカーの大きさデータ(s):正規分布から抽出した50個の乱数(c)->発生した乱数の絶対値を取った後、100倍
# yデータ(d):xデータ+10*正規分布から抽出した50個の乱数
data = {'a':np.arange(0, 50),
       'b':np.random.randint(0, 50, 50),
       'c':np.random.randn(50)}

data['d'] = data['a'] + 10*data['c']
data['s'] = np.abs(data['c'])*100
```

```python
#データ確認 : yに該当するデータ
data['d']
```

```python
# データ確認 :マーカーサイズに該当するデータ
data['s']
```

```python
# dataパラメータを通じて伝えたオブジェクトで
# x, y, s, c パラメータに該当する値に使用しようとする Key を文字で伝達
# plt.scatter('x', 'y', s(size)='s', c(color)='c', data=dict)
plt.scatter('a', 'd', s='s', c='b', data=data)
plt.show()
```

### Bar Chart

- 様々な値の相対的な大きさを比較するのに適した表現法であり、円形チャートよりさらに一般的に使われる。
- plt.bar(x, y)
    - x データがカテゴリ値の場合には bar, barh で Bar Chart 視覚化表現
- 横方向にBar Chartを描くにはbarh
    - barhチャートの注意点 : x軸が一般的にy軸の位置を意味

```python
x = ['가', '나', '다']
y = [2, 10, 5]
plt.bar(x, y)
```

```python
plt.bar(x, y)
plt.title('bar chart')
plt.xlabel('x축')
plt.ylabel('y축')
plt.xticks(x, ['김', '이', '박'])
plt.show()
```

```python
bar_list = plt.bar(x, y)
bar_list[0].set_color('r')
bar_list[1].set_color('g')
plt.title('bar chart')
plt.xticks(x, ['김', '이', '박'])
plt.xlabel('x축')
plt.ylabel('y축')
plt.show()
```

#### 横型の棒グラフ

- plt.barh(y, x)

```python
bar_list = plt.barh(x, y)
bar_list[0].set_color('r')
bar_list[1].set_color('g')
plt.title('bar chart')
plt.xlabel('x축')
plt.ylabel('y축')
plt.show()
```

### Pie chart

- 全データに各部分が占める割合を視覚的に理解しやすく表現する。
- plt.pie(ratio, labels=, colors=, autopct=, shadow=T/F)
    - ratio : 各ラベルの比率
    - label : データのカテゴリ名
    - color: ラベル別の色
    - autopct: 比率出力フォーマット
    - shadow:影を指定

```python
plt.pie([15, 30, 45, 10],
        labels=['소고기', '돼지고기',  '닭고기', '오리고기'],
       colors=['yellowgreen', 'gold', 'lightskyblue', 'lightcoral'],
       autopct='%0.1f%%', shadow=True)
plt.axis('equal') # 円の形を設定
plt.show()
```

### Histogram

- データ分布の特徴と異常値の有無を確認するのに容易である.
    - データの分布
        - データ値の全般的に広がっている形.
        - データの数が多い時、データを把握するためにどの値の近くにどんな形で集まっているのか理解しやすい.
- データに異常値が存在する場合、異常なHistogramが生成され、異常値を除去する作業が必要だ.
    - 例)全般的なデータ分布が0~1の間にあるが、一つのデータが100の場合、正常なHistogramが描かれない.

    - 資料値が持てる範囲をいくつかの区間に分け、各区間に該当する値の数字あるいは相対的頻度を計算して視覚化する.
    - plt.hist(data, bins)
        - 一般的な場合、Histogramの区間を分ける際に同一間隔で分けるため、区間(bins)の数が大切である.
        - bins
            -ヒストグラムの区間数を表す値である.
            - 区間数が小さすぎると(=間隔が広すぎると)細部分布の把握が難しく、区間数が多すぎると全体的な分布の把握が難しい.

```python
# 正規分布から無作為に1000のデータ抽出
np.random.seed(200)
x = np.random.randn(1000)
plt.hist(x, bins=50) # None vs 300 vs 100 vs 50
plt.show()
```

### Box-plot

- box-and-whisker plot
- 最大値、最少値、四分位数(Q1、Q2、Q3)を使用し、資料の測定値の分布と劇団の値について把握する。
    - IQR(Inter Quartile Range):Q3-Q1
    - 最大値:Q1+1.5 x IQRの範囲内にある最大値
    - 最小値:Q3-1.5 x IQRの範囲内にある最大値
    - ボックス(Box):Q1~Q3の値
    - n個のデータがある時の四分位数
        - 中央値(Q2):(n+1)/2
        - 四分衛手
            - 集荷資料を大きさ順に並べて四等分する尺度。
            - (Q2+1)/2を基準に低い方で、該当位置にある数を下四分位数(Q1)、大きい方に位置した数をウィサブンの玉樹(Q2=n+1-値)
            - データ個数が奇数のときは前後の値の平均を使う。
- 二つ以上の資料集団を比較する時に有用だ。
- 当該範囲から外れる値は異常値と判定して表記する。

```python
# 平均70、標準偏差25人の正規分布で200社の標本を抽出
data = np.random.normal(70, 25, 200)
data
```

```python
plt.hist(data)
```

```python
plt.boxplot(data)
plt.show()
```
