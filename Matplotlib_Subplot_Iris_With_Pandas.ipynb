## 1.数値をデータで、ソクソンミョンをColumnにはDataFrameを生成

```python
df = pd.DataFrame(dataset['data'], columns=dataset['feature_names'])
df.head()
```

## 2.targetデータとtarget_namesデータを結合して品種の名前のデータをColumnに追加

```python
# 각 수치형 target을 문자형/명목형 데이터로 변형하여 컬럼으로 결합
# 0 -> setosa, 1 -> versicolor, 2 -> virginica
# [(k,v)]
# zip([1,2,3], ['a','b','c'])
# zip은 튜플로 만들어주는 함수.
species_map = dict(zip(range(3), dataset['target_names']))
species_map
```

```python
# 新しいColumnのspecies生成
# ColumnのIndexing = リスト
df['species'] = [species_map[i] for i in dataset['target']]
[species_map[i] for i in dataset['target']]
```

```python
# 最終データの確認
df
```

### Pie Plot
- 全体で特定の範疇が占める割合と割合を把握するために使用
- pandas.DataFrame/Series.plot(kind="'pie")
    - Seriesオブジェクト:一つの円形チャートを生成
    - DataFrameオブジェクト: 各列ごとに一つずつ円形のチャートを生成
- pandas.DataFrame/Series.plot.pie(y=None, figsize=None, subplots=False)
    - y : ひとつのコラムに対して円形チャートを生成するときに使用するコラム名
    - figsize: 生成するfigureオブジェクトサイズ(tuple)
    - subplots: y値で一つのコラムを使うのではなく、全体のdfを使うときにTrueに指定

```python
# すべての品種のデータの平均
df.head()
# groupbyもしくはpivot_tableを利用して品種別にデータ平均。
mean_by_species = df.groupby('species').mean()
mean_by_species
mean_by_species2 = df.pivot_table(index='species', values=
                                  ['sepal length (cm)','sepal width (cm)','petal length (cm)','petal width (cm)'],aggfunc='mean')
mean_by_species2
```

```python
mean_by_species.plot(kind='pie', subplots=True, layout=(2,2), legend=False)
```

```python
mean_by_species.plot(kind='pie', y='sepal length (cm)', legend=False,
                    label='', title='< sepal length >')
```

### イメージの保存

- matplotlib.pyplot.savefig(fname、bbox_inches=None、pad_inches=0.1)
    - fname:別途の経路を指定しなければ、現在の作業経路にイメージファイルを保存する. png、pdf形式をサポートし、jpg形式は別の変換作業が必要である。
        - JPG ファイルに変換
            - import Image
            - Image.open('file.png')save('file2.jpg'、'JPEG')

- bbox_inches : ファイルとして保存される領域を設定 ('tight' : 視覚化されたfigureに合わせてイメージファイルを生成する. )
- pad_inches:保存されるイメージファイルの余白を指定する.

```python
mean_by_species.plot(kind='pie', y='sepal length (cm)', legend=False, label='', title='< sepal length >')
plt.savefig('pie1.png')
plt.savefig('pie2.png', bbox_inches='tight')
plt.savefig('pie3.png', bbox_inches='tight', pad_inches=5)
```

### Histogram

- pandas.DataFrame/Series.plot(kind='hist')
    - Seriesオブジェクト:一つのヒストグラムを生成
    - DataFrameオブジェクト: 各列ごとに一つずつHistogramのチャートを生成, subplots=True & layout = (n,m)
- pandas.DataFrame/Series.hist(bins=10)
    - bins: Histogram区間の数を表す値で区間の数が小さすぎると(=間隔が広すぎると)細部の分布把握が難しく、区間の数が多すぎると全体的な分布把握が難しい。

```python
# すべての品種を対象とする各属性に対するヒストグラム
# 一番目の方法:dfplot(kind='hist')
df.plot(kind='hist')
```

```python
df.plot(kind='hist', subplots=True)
```

```python
df.plot(kind='hist', subplots=True, layout=(2,2))
```

```python
df.plot(kind='hist', subplots=True)
plt.suptitle('Iris Histogram', fontsize=20)
plt.show()
```

```python
# 二番目の方法:bins=10、binsは、区間を何で分ける時に使う関数
df.hist()
```

```python
df.hist(bins=3)
```

```python
# ver1. df.plot(kind='hist')
for spec in df['species'].unique(): # 'setosa', 'versicolor', 'virginica' を順番にspecに割り当てる.
    tmp = df[df['species'] == spec] # 'setosa', 'versicolor', 'virginica' を順番にtmpに割り当てる.
    tmp['petal length (cm)'].plot(kind='hist') # 該当名前のpetal length (cm)でHistogramを描く.
     
plt.legend()
plt.suptitle('종별 꽃잎 길이')
plt.show()
```

```python
# ver2. plot(kind='hist') - ver1の問題解決
for spec in df['species'].unique():
    tmp = df[df['species']==spec]
    # label -> legendで表示される名前、gridの格子線を表示するかどうか
    tmp['petal length (cm)'].plot(kind='hist', label=spec, grid=True)

plt.legend()
plt.suptitle('종별 꽃잎 길이')
plt.show()
```

### scatter-plot

- 二種類の変数含めて適用することができる。
- 二つの変数との関係について分散と分布を通じて把握できる。
- pandas.DataFrame/Series.plot(kind='scatter', x=None, y=None)
    - x : 横(x)に使われるデータとColumn名
    - y:縦に使われるデータとColumn名
- pandas.DataFrame/Series.plot.scatter(x, y, s=None, c=None)
    - x : 横(x)に使われるデータとColumn名
    - y:縦に使われるデータとColumn名
    - s : 各点の大きさ
    - c:各点の色

```python
df.drop('idx', axis=1, inplace=True)
```
