# OuterJoinProgram

- 様々なプラットフォームから取り寄せた実績を合わせるためのプログラムです.
- まずシート数とGroupby数をセットします.
- ボックスから個数を選択し、"シート数"と"gropby数"ボタンを押すと、セットした通りにウィンドウが変わります.
- また、スピンボックスに1という値がセットされます.
- ファイル照会ボタンを押すと現在ファイルの1番シートにあるColumnが出ます.
- リストにあるColumn名をクリックし、矢印ボタンを押すとセットされます.
- 矢印の方向を押すと、Columnが移動するのを見ることができます.
- セッティングを終えたColumnは実際に使用され、真ん中に残ったColumnは後で自動的に消えます.
- もう使用Column保存ボタンを押すと基準Columnと使用Columnが入力されます.
- 次にスピンボックスで上の矢印を押すと数字が変わります.
- ファイル照会ボタンを押せば、二回目のシートを照会することになります.
- 使用コラムの保存ボタンを押してセットすればいいです.
- このようなやり方で望んでいたシート全部をセットしてくれたら全てのセットが終わります.
- 基準Columnで使うColumn名を全部同じに変えてくれて使用Column名も識別できるように変換をしなければなりません.
- 基準列と使用列の名称は、各番号データに合った名前をコピーして入れます.データは区切り文字を","にします。
- ","を入れて区分し、Column名を順に変えます.
- 上から下記のような形に変えて記入し、右側のColumn修正ボタンを押します.
- それではセッティングを完了したという意味で背景の色が白色に変わります.
- 同じ原理ですべてのColumnを修正してくれます.
- 基準Columnの手順まで合わせる必要はありません.
- 基準Columnが配置された順に名前を同じように作ります.(大小文字区分)
- もうファイル保存ボタンを押すと、Group byオプションが適用されたままMerge作業が実行されます.
- このようにして作業したファイルは Raw データを持ち込んだフォルダに xxx-result.xlsx というファイルで保存されます.
- 最後にデータがうまく合わされているか確認します。
- データがExportされていなければ、途中順やColumn名の記入などにエラーがあったわけです。
