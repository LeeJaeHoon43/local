```python
import pandas as pd
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
```

### df.plot()

- Pandasデータタイプのplotはmatplotlibを内部でインポートして使用
- plotチャート種類: kindパラメータ値で配信
    - line(基本)
    - bar
    - pie
    - hist
    - box
    - scatter

```python
# 株価のデータと似た流れを見せるサンプルを生成
# value:標準正規分布で100 x 3の構造の標本を抽出
# Column 3つ : A,B,C
# index : pd.date_range(start, [end], periods=n)
np.random.seed(111)
data = np.random.randn(100, 3)
data
```

```python
df = pd.DataFrame(data, 
                  #2018-08-07から100日間の日付を生成
                 index=pd.date_range('8/7/2019', periods=100),
                 columns=['A', 'B', 'C']).cumsum() # cumsum()은 누적합이다.
df 
```

```python
# lineplot(基本): matplotのpltではなくpandasのDataFrameにすぐに使用することもできる。
df.plot()
```

```python
#ライブラリで提供する基本データ
import seaborn as sns
```

```python
# データ保存
iris = sns.load_dataset('iris')
```

```python
iris.shape
```

```python
iris.head()
```

```python
# plotで韓国語のfontを指定
font_name=matplotlib.font_manager.FontProperties(fname='C:/Windows/Fonts/malgun.ttf').get_name()
matplotlib.rc('font', family = font_name)
```

```python
# sepal_lengthの20個のサンプルについてbar graphで視覚化する。
# series.plot()
iris.sepal_length[:20].plot(kind='bar', rot=0)
plt.title('샘플 20개에 대한 꽃받침의 길이')
plt.xlabel('샘플번호')
plt.ylabel('꽃받침의 길이')
plt.show()
```

```python
# サンプルデータ(5個)の各プロパティ値(4項目)に対してbar graphを生成
iris[:5].plot(kind='bar', rot=0) # plot.bar
plt.title('샘플 5개에 대해 4가지 속성값 비교')
plt.xlabel('sample')
plt.ylabel('cm')
plt.ylim(0, 8)
plt.show()
```

```python
iris
```

```python
# 各種類別に4つのプロパティ値の平均=>各属性が種類別に違いがあるのかどうか見るために
iris.pivot_table(index='species', values=['sepal_length', 'sepal_width', 'petal_length', 'petal_width'], aggfunc='mean')
```

```python
# .groupby()を使うタイプ
df2 = iris.groupby(iris.species).mean()
df2.columns.name = 'feature'
df2
```

```python
# bar graphで視覚化
# インデックス別にColumn値を棒のグラフで視覚化
df2.plot(kind='bar', rot=0)
plt.title('품종별 속성값 평균')
plt.xlabel('품종')
plt.ylabel('cm')
plt.ylim(0, 9)
plt.show()
```

```python
df2.T.plot(kind='bar', rot=0)
plt.show()
```

```python
# 各属性別にデータ値に対する技術統計の値と異常値を確認。 => box-plot
# df.plot(kind="'box") => Columnを基準に記述統計,異常値を確認
iris.plot(kind='box')
plt.show()
```

```python
# vesicolor
iris[iris['species']=='versicolor'].plot(kind='box')
```

```python
# verginica
iris[iris['species']=='virginica'].plot(kind='box')
```

```python
# 品種ごとに属性別のboxplotを生成 => 一つのplot内部に複数のboxplot.
# iris.plot (kind="'box") # box plot => Column (数値データ)
iris.boxplot(by='species')
plt.tight_layout(pad=3) 
plt.show()
```
