# 映画の評点を分析する実習

```python
import pandas as pd
```

## 1.映画の評点データを積載して結合

```python
#ユーザーの情報を積載
users = pd.read_csv('data/movie_users.csv')
users.drop('Unnamed: 0',axis=1, inplace=True)
users.head()
```

```python
# 映画の評点情報を積載
ratings = pd.read_csv('data/movie_ratings.csv')
ratings.drop('Unnamed: 0', axis=1, inplace=True)
ratings.head()
```

```python
movies = pd.read_csv('data/movie_details.csv')
movies.drop('Unnamed: 0', axis=1, inplace=True)
movies.head()
```

```python
print('''사용자수 : {}
영화수 : {}
평점수 : {}'''.format(len(users), len(movies), len(ratings)))
```

### 1.データ結合
#### 1-1) 使用者データや映画の評点データをユーザ・アイディColumnを基準に共通に存在するデータのみ結合する。

```python
#ユーザー情報のColumndの名を確認
users.columns
```

```python
#評点入力の情報のColumnの名を確認
ratings.columns
```

```python
# users, ratingsの配列構造を確認
print(users.shape)
print(ratings.shape)
```

```python
#data1変数にusers、ratingsの順でpd.merge()
data1 = pd.merge(users, ratings)
#実行後にdata1.shapeに配列の構造を確認
data1.shape
```

#### 1-2)の結果に映画のハンドルネームColumnを基準に映画の情報の追加する。

```python
data1.head()
```

```python
# 映画データのmovies
movies.head()
```

```python
# data1とmoveisをMergeしてdata2変数に保存して映画情報を追加する.
data2 = pd.merge(data1, movies)
data2.head()
```

```python
data2.shape
```

### 2.人気のある映画10つみつけこと

- 基準 : 評点が4.5以上で評点の個数が2000個以上の映画

```python
# 評点が高い順に10個を整列
data2.sort_values(by='평점', ascending=False).head(10)
```

```python
# Row(index) : 映画のタイトル
# Column(values) : 映画の評点
# 集計関数(aggfuns) : 平均
data2.pivot_table(index='제목', values='평점', aggfunc='mean')
```

```python
data2.pivot_table(index='제목', values='평점', aggfunc='mean').sort_values(by='평점', ascending=False)
```

```python
# 評点を基準に整列し、評点だけを出力するのではなく、入力された評価の数も一緒に出力する。
# Row : タイトル
# Value(Column) : 評点
# 集計関数 : mean, count
result = data2.pivot_table(index='제목', values='평점', aggfunc=['mean', 'count'])
```

```python
result.head()
```

```python
# result에 대해 컬럼의 계층색인 재배열(평점을 카테고리로, mean, count를 하위 컬럼으로)
result2 = result.swaplevel(1, 0, axis=1)
```

```python
result2.head()
```

```python
# 評点がmeanで同一の場合には評点のcount順に整列させる。
result2.sort_values(by=[('평점', 'mean'), ('평점', 'count')], ascending=[False, False])
```

```python
# 条件索引を利用してresult2で評点の平均が4.5以上でありながら評点の個数が500個以上の映画だけを残している。
# 二重的なColumnはresult2['評点']['mean']のように二重的なリストで照会する。
final_result = result2[(result2['평점']['mean']>=4.5) & (result2['평점']['count'] >= 500)]
```

```python
# final_resultを評点の順に、評点が同点ならcount順に並べる。
final_result.sort_values(by=[('평점', 'mean'), ('평점', 'count')], ascending=[False, False])
```
