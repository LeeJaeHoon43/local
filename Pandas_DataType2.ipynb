```python
import pandas as pd
import numpy as np
```

### 算術演算

- dataframe とスカラ値算術演算
- dataframe と series 間の算術演算
- dataframe間の算術演算
    - Column, Rowインデックスを基準に演算遂行
    - 共通が存在しない場合NaN返還
-fill_value値を通じてNaNではなく値に代替可能



- 演算の種類
    - +, add()メソッド
    - -, sub()メソッド
    - * , mul()メソッド
    - %
    - //

```python
df1 = pd.DataFrame(np.arange(9).reshape(3, 3), index=list('abc'),
                  columns = ['서울', '경기', '인천'])
df1
```

```python
df2 = pd.DataFrame(np.arange(20).reshape(4, 5), index=list('abcd'), 
                   columns=['서울', '경기', '인천', '세종', '강원'])
df2
```

```python
df1 + df2
```

```python
df1.add(df2, fill_value=0)
```

```python
df1 - df2
```

```python
df1.sub(df2, fill_value=0)
```

```python
df1 * df2
```

```python
df1.mul(df2, fill_value=1)
```

```python
df1 / df2
```

```python
df1.divide(df2, fill_value=1)
```

```python
df1 % df2
```

### DataFrame と Series 間の演算

- 基本的な動作はSeriesオブジェクトのインデックスをDataFrameオブジェクトのColumn indexとMappingしてBroadCastingと類似に演算を遂行
- オブジェクト間の共通のインデックスではなく対象はNaN値で代入
- 関数を使用して演算を修行する際にはaxisパラメータを通じて演算を適用する軸を指定(0:行、1:列)
- Columnとの本数が一致しなければならない。

- 演算の種類
    - +, add()
    - -, sub()
    - *, mul()

```python
df = pd.DataFrame(np.arange(12).reshape(3, 4), index=[2010, 2011, 2012],
                 columns=list('abcd'))
df
```

```python
s1 = df.iloc[0]
s1
```

```python
df
```

```python
df + s1
```

```python
df2 = pd.DataFrame(np.zeros((4, 5)), columns=list('abcde'))
df2
```

```python
s2 = pd.Series(np.arange(5))
s2
```

```python
df2.sub(s2)
```

```python
df2.sub(s2, axis=0)
```

```python
df
```

```python
s3 = pd.Series([3, 3, 3], index=list('ace'))
s3
```

```python
df - s3
```

```python
df.sub(s3)
```
